```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score,
confusion_matrix

# 1. Chargement du jeu de données
df = pd.read_csv('maroc_bank_transactions.csv')

# 2. Nettoyage et sélection des variables
df_clean = df.drop(columns=['transaction_id', 'time', 'type_fraud'])
X = df_clean.drop(columns=['status'])
y = df_clean['status']

# 3. Identification des types de variables
categorical_cols = ['bank_name', 'city', 'card_type']
numerical_cols   = ['amount_MAD']

preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'),
categorical_cols)
])
X_preprocessed = preprocessor.fit_transform(X)

# 4. Découpage Train/Test (80%/20%) avec stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_preprocessed, y, test_size=0.2, random_state=42, stratify=y)

# 5. Entraînement du modèle avec pondération des classes
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

# 6. Prédictions et évaluation
y_pred = model.predict(X_test)
print(f"Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
